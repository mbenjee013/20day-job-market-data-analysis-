```python
import pandas as pd
```

1. Created a dataframe---> df
2. Remove unnecessary columns---> id, closing_date, expire_status, author, key_words and newsletter
3. Drop duplicates---> which reduced the total number of jobs to 1522
4. rename columns---> mode=employment type, post=description and category=industry

```python
df = pd.read_csv('pezantchito_dataset.csv')
df.drop(columns=['id', 'closing_date', 'expire_status', 'author', 'key_words', 'newsletter'], inplace=True)
df.drop_duplicates(inplace=True)
df.rename(columns={"mode":"employment type", "post":"description", "category":"industry"}, inplace=True)
```

1. Coverted all values to string and lowercase
2. Created a mapping dictionary to clean inconsistencies
3. Replced messy values with standardised employment types
4. Come up with 12 unique employment types

```python
uniqueEmploymentType={
    "-----":"undefined", "----":"undefined", "--":"undefined", "---":"undefined",
    "full-time (monday-friday)":"fulltime", "full time":"fulltime", "full-time":"fulltime",
    "intern":"internship", "internship(paid)":"internship","volunteer":"voluntary","short-term":"shortterm",
    "contract (renewal)":"contract", "contract based":"contract", 
    "contract(1yr)":"contract", "contract(2yrs)":"contract", "contract(12 month)":"contract",
    "contract(4yrs)":"contract", "contract(renewable)":"contract", "contract(2 years)":"contract",
    "contract(3 years)":"contract", "contract (1yr)":"contract", "contract(6 months)":"contract", 
    "contract(3yrs)":"contract", "contract (renewal), full-time (38 hours 45 minutes/week).":"contract",
    "part time":"parttime", "part-time":"parttime", "full+part time":"parttime","fixed-term / permanent":"permanent",
    "temporary (6 months)":"temporary", "temporally":"temporary","temporary contract":"temporary",
    "fixed term":"fixedterm", "fixed-term":"fixedterm", "fixed contract":"fixedterm",
    "fixed term contract (8 months)":"fixedterm","fixed term contract(12 months)":"fixedterm",
    "commission-based":"commission based", "commission":"commission based"
}

df['employment type']=df['employment type'].str.lower().replace(uniqueEmploymentType)
df['employment type'].unique()
df
```
